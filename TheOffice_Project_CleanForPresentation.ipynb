```python
from scipy import stats

writers = office_df_clean['writer'].unique()

sentiment_by_writer = [office_df_clean[office_df_clean['writer'] == writer]['sentimentAnalysis_score'] for writer in writers]


f_statistic, p_value = stats.f_oneway(*sentiment_by_writer)

print(f"One-Way ANOVA F-statistic: {f_statistic}")
print(f"One-Way ANOVA p-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print(f"Reject the null hypothesis (p < {alpha}). There is a significant difference in mean sentiment analysis score among different writers.")
else:
    print(f"Fail to reject the null hypothesis (p >= {alpha}). There is no significant difference in mean sentiment analysis score among different writers.")

mean_sentiment_by_season = office_df_clean[office_df_clean['season'].isin([1, 9])].groupby('season')['sentimentAnalysis_score'].mean()

plt.figure(figsize=(8, 6))
mean_sentiment_by_season.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title("Mean Sentiment Analysis Score by Season (1 vs 9)")
plt.xlabel("Season")
plt.ylabel("Mean Sentiment Analysis Score")
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

office_df_main = office_df_clean[['season', 'episode_name', 'director', 'character', 'text', 'text_w_direction', 'imdb_rating', 'sentimentAnalysis_score', "writer"]]

features = office_df_main[['season', 'writer', 'character']]
label = office_df_main['sentimentAnalysis_score']

features = pd.get_dummies(features, columns=['season', 'writer', 'character'], drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=2025)

model_rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=2025) # Added max_depth
model_rf.fit(X_train, y_train)

y_pred_rf = model_rf.predict(X_test)

plt.figure(figsize=(8, 8))
plt.scatter(y_test, y_pred_rf, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--k', label="Correct prediction")
plt.xlabel('True Sentiment Analysis Score')
plt.ylabel('Predicted Sentiment Analysis Score')
plt.title("Random Forest Regression: Real vs. Predicted Sentiment Analysis Score (max_depth=5)")
plt.axis('tight')
plt.legend()
plt.tight_layout()
plt.show()


print('Random Forest Regression Metrics (Predicting Sentiment Score, max_depth=5):')
print('Mean squared error: ', mean_squared_error(y_test, y_pred_rf))
print('Mean absolute error: ', mean_absolute_error(y_test, y_pred_rf))
print('Mean absolute percentage error: ', mean_absolute_percentage_error(y_test, y_pred_rf))
print('R squared: ', r2_score(y_test, y_pred_rf))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
import pandas as pd
import matplotlib.pyplot as plt

office_df_main = office_df_clean[['season', 'episode_name', 'director', 'character', 'text', 'text_w_direction', 'imdb_rating', 'sentimentAnalysis_score', "writer"]]

features = office_df_main[['season', 'writer',]]
label = office_df_main['sentimentAnalysis_score']

features = pd.get_dummies(features, columns=['season', 'writer'], drop_first=True)

poly = PolynomialFeatures(degree=2, include_bias=False)
features_poly = poly.fit_transform(features)

X_train_poly, X_test_poly, y_train, y_test = train_test_split(features_poly, label, test_size=0.2, random_state=2025)

model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)

y_pred_poly = model_poly.predict(X_test_poly)

plt.figure(figsize=(8, 8))
plt.scatter(y_test, y_pred_poly, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--k', label="Correct prediction")
plt.xlabel('True Sentiment Analysis Score')
plt.ylabel('Predicted Sentiment Analysis Score')
plt.title("Polynomial Regression: Real vs. Predicted Sentiment Analysis Score (degree=2)")
plt.axis('tight')
plt.legend()
plt.tight_layout()
plt.show()

print('Polynomial Regression Metrics (Predicting Sentiment Score, degree=2):')
print('Mean squared error: ', mean_squared_error(y_test, y_pred_poly))
print('Mean absolute error: ', mean_absolute_error(y_test, y_pred_poly))
print('Mean absolute percentage error: ', mean_absolute_percentage_error(y_test, y_pred_poly))
print('R squared: ', r2_score(y_test, y_pred_poly))
```

```python
from sklearn.svm import LinearSVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

office_df_main = office_df_clean[['season', 'episode_name', 'director', 'character', 'text', 'text_w_direction', 'imdb_rating', 'sentimentAnalysis_score', "writer"]]

features = office_df_main[['season', 'writer', 'character']]
label = office_df_main['sentimentAnalysis_score']

features = pd.get_dummies(features, columns=['season', 'writer', 'character'], drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=2025)

model_svm = LinearSVR(random_state=2025, max_iter=10000) # Increased max_iter

model_svm.fit(X_train, y_train)

y_pred_svm = model_svm.predict(X_test)

plt.figure(figsize=(8, 8))
plt.scatter(y_test, y_pred_svm, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--k', label="Correct prediction")
plt.xlabel('True Sentiment Analysis Score')
plt.ylabel('Predicted Sentiment Analysis Score')
plt.title("Linear SVR: Real vs. Predicted Sentiment Analysis Score")
plt.axis('tight')
plt.legend()
plt.tight_layout()
plt.show()


print('Linear SVR Metrics (Predicting Sentiment Score):')
print('Mean squared error: ', mean_squared_error(y_test, y_pred_svm))
print('Mean absolute error: ', mean_absolute_error(y_test, y_pred_svm))
print('Mean absolute percentage error: ', mean_absolute_percentage_error(y_test, y_pred_svm))
print('R squared: ', r2_score(y_test, y_pred_svm))
```
